```python
import gurobipy as gp
from gurobipy import GRB

q1_s1 = gp.Model("Q1_Nucleolus_method")

x1 = q1_s1.addVar(name="x1", lb=10000)
x2 = q1_s1.addVar(name="x2", lb=0)
x3 = q1_s1.addVar(name="x3", lb=0)
w = q1_s1.addVar(name="w", lb=-GRB.INFINITY)

q1_s1.setObjective(w, GRB.MAXIMIZE)

q1_s1.addConstr(x1 + x2 + x3 == 30000, "c1")
q1_s1.addConstr(x1 - w >= 10000, "c2")
q1_s1.addConstr(x2 - w >= 0, "c3")
q1_s1.addConstr(x3 - w >= 0, "c4")
q1_s1.addConstr((x1 + x2) - w >= 20000, "c5")
q1_s1.addConstr((x1 + x3) - w >= 30000, "c6")
q1_s1.addConstr((x2 + x3) - w >= 0, "c7")

q1_s1.optimize()

print(f"Player 1 = {x1.X}, Player 2 = {x2.X}, Player 3 = {x3.X}, Minimum excess = {w.X}")
```

```python
q1_s2 = gp.Model("second_stage_q1")

w = q1_s2.addVar(name="w", lb=-GRB.INFINITY)
x2 = q1_s2.addVar(name="x2")
x3 = q1_s2.addVar(name="x3")

q1_s2.setObjective(w, GRB.MAXIMIZE)

q1_s2.addConstr(x2 - w >= 0, "c1")
q1_s2.addConstr(x3 - w >= 0, "c2")
q1_s2.addConstr(x2 + x3 == 10000, "c3")
q1_s2.addConstr(x2 >= 0, "c4")
q1_s2.addConstr(x3 >= 0, "c5")

q1_s2.optimize()

print(f"Player 2 = {x2.X}, Player 3 = {x3.X}, Minimum excess = {w.X}")
```

# **Question 2**

```python
import gurobipy as gp
from gurobipy import GRB

q2_s1 = gp.Model("Nucleolus_Approach")

x1 = q2_s1.addVar(name="x1", lb=0)
x2 = q2_s1.addVar(name="x2", lb=0)
x3 = q2_s1.addVar(name="x3", lb=0)
w = q2_s1.addVar(name="w", lb = -GRB.INFINITY)

q2_s1.addConstr(x1 - w >= 0, "c1")
q2_s1.addConstr(x2 - w >= 0, "c2")
q2_s1.addConstr(x3 - w >= 0, "c3")
q2_s1.addConstr(x1 + x2 - w >= 0, "c4")
q2_s1.addConstr(x1 + x3 - w >= 0, "c5")
q2_s1.addConstr(x2 + x3 - w >= 100000, "c6")
q2_s1.addConstr(x1 + x2 + x3 == 200000, "c7")

q2_s1.setObjective(w, GRB.MAXIMIZE)

q2_s1.optimize()

print(f"Player 1 = {x1.X}, Player 2 = {x2.X}, Player 3 = {x3.X}, Minimum excess = {w.X}")
```

```python
q2_s1 = gp.Model("second_stage_q2")

x2 = q2_s1.addVar(name="x2", lb=0)
x3 = q2_s1.addVar(name="x3", lb=0)
w = q2_s1.addVar(name="w", lb=-GRB.INFINITY)

q2_s1.addConstr(x2 - w >= 0, "c1")
q2_s1.addConstr(x3 - w >= 0, "c2")
q2_s1.addConstr(x2 + x3 == 150000, "c3")
q2_s1.setObjective(w, GRB.MAXIMIZE)

q2_s1.optimize()

print(f"Player 2 = {x2.X}, Player 3 = {x3.X}, Minimum excess = {w.X}")
```
